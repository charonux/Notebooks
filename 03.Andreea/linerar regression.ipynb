```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import optimizers
from tensorflow.keras import losses
```

```python
 x= np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4], dtype=float)
 y= np.array([-14, -11 ,-8 ,-5 ,-2 ,1 ,4 ,7 ,10], dtype=float)
# x = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
# y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
model = models.Sequential()
model.add(layers.Dense(units=1, input_shape=[1]))
```

```python
model.compile(optimizer=optimizers.SGD(lr=0.01), loss=losses.mean_squared_error)
model.summary()
```

```python
model.fit(x,y,epochs=500)
```

```python
for layer in model.layers:
   param = layer.get_weights()

print('Weight =', param[0], 'Bias =', param[1])
```
