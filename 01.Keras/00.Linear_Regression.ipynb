```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.callbacks import TensorBoard
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import optimizers
from tensorflow.keras import losses
from tensorflow.keras import metrics
```

# Preparing the data

- f(x)=2x-1

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

# Build the model

```python
model = models.Sequential()
model.add(layers.Dense(units=1, input_shape=[1]))

tensorboard = TensorBoard(log_dir='/notebooks/99.logs')
```

# Model compilation

```python
model.compile(optimizer=optimizers.SGD(lr=0.01), loss=losses.mean_squared_error)

model.summary()
```

# Train the model

```python
model.fit(xs, ys, epochs=500, callbacks=[tensorboard])
```

# Generate prediction on new data

```python
print(model.predict([10.0]))
```

- Extract liniar function parameters f(x)= x*weight + bias

```python
for layer in model.layers:
   param = layer.get_weights()

print('Weight =', param[0], 'Bias =', param[1])
```
